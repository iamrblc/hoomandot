```python
import pandas as pd
```

```python
# Create empty dataframe with LUMO's names
channels = pd.DataFrame(columns=['idx',                 # channel index
                                 'src_node_id',         # which tile is the source on
                                 'src_idx',             # source index
                                 'src_optode_name',     # source name (A, B, C)
                                 'src_coords_3d',       # source coordinates (i omit 2D)
                                 'src_wl',              # wavelength
                                 'det_node_id',         # which tile is the detector on
                                 'det_idx',             # detector index
                                 'det_optode_name',     # detector name (same as det_idx)
                                 'det_coords_3d'],      # detector coordinates (i omit 2D)
                                 index=range(3456))

length = len(channels)
```

```python
# idx is the channel number (1-3456)
channels['idx'] = channels.index + 1
channels
```

```python
# Wavelegth should alternate between 735 an 850 in chunks of 144
channels['src_wl'] = ['735' if (i // 144) % 2 == 0 else '850' for i in range(length)]
channels
```

```python
# Source index moves up one every 48 channels
channels['src_idx'] = [(i // 48) + 1 for i in range(length)]
```
